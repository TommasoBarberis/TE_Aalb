# Consensi analysis

```python
# library import
from Bio import SeqIO
import pandas
import plotly.express as px
import plotly.graph_objects as go
```

## RepeatModeler2

```python
seq_lens = []
for seq_record in SeqIO.parse("data/RM2_consensi.fa", "fasta"):
    seq_id = seq_record.id
    seq_lens.append(len(seq_record.seq))

fig = px.histogram(seq_lens, labels={'value':'sequence length'})
fig.show()
```

```python
print(len(seq_lens))
print(max(seq_lens))
print(min(seq_lens))
```
